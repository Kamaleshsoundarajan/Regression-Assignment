```python
#Importing pandas library

import pandas as pd
```

```python
#upload the dataset file

dataset=pd.read_csv("insurance_pre (1).csv")
```

```python
dataset
```

```python
#Converting the categorical data into Numerical Data
#Eliminating the dummies from the columns

dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#list the name of the columns

dataset.columns
```

```python
#input split

independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
#output split

dependent=dataset[['charges']]
```

```python
dependent
```

```python
#train & test set split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation

from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error',splitter='random')
regressor.fit(x_train,y_train)
```

```python
#View the decision tree model 

import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#Evaluating the model

y_pred=regressor.predict(x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
